```python
import tensorflow as tf
import os

# 1. Point to your folders
image_dir = '../data/images/'
model_dir = '../models/'

# 2. Load and format the images
print("Loading images from your data folder...")
# We split the data: 80% to train the AI, 20% to test if it's actually learning
train_ds, val_ds = tf.keras.utils.image_dataset_from_directory(
    image_dir,
    image_size=(224, 224), # Resize all images to standard AI size
    batch_size=32,
    validation_split=0.2,  
    subset="both",
    seed=123
)

class_names = train_ds.class_names
print(f"\nAwesome! The AI found {len(class_names)} different disease classes.")

# 3. Build the Brain (Using MobileNetV2)
print("\nBuilding the AI Model...")
base_model = tf.keras.applications.MobileNetV2(
    input_shape=(224, 224, 3),
    include_top=False,
    weights='imagenet'
)
base_model.trainable = False # Freeze the base brain, we only train the new output layer

# Attach our custom plant disease layer to the top
model = tf.keras.Sequential([
    base_model,
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dense(len(class_names), activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 4. Train the Model
print("\nStarting training for 3 Epochs (rounds)...")
history = model.fit(train_ds, validation_data=val_ds, epochs=3)

# 5. Save the Brain
save_path = os.path.join(model_dir, 'plant_disease_model.keras')
model.save(save_path)
print(f"\nSuccess! Model saved to: {save_path}")
```
