```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# 1. Find the exact name of your downloaded dataset
data_folder = '../data/tabular/'
dataset_name = 'crop_yield.csv'
file_path = os.path.join(data_folder, dataset_name)

print(f"Loading data from: {file_path}\n")

# 2. Load the weather and yield data
df = pd.read_csv(file_path)

# 3. Show the first 5 rows
print("Here is what your data looks like:")
display(df.head())

# 4. Show a simple correlation heatmap
plt.figure(figsize=(8, 5))
numeric_data = df.select_dtypes(include=['number']) # We only look at numbers, not text
sns.heatmap(numeric_data.corr(), annot=True, cmap='Blues', fmt=".2f")
plt.title("Correlation: Weather vs Crop Yield")
plt.show()
```
