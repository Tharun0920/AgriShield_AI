```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the weather and yield data
# Note: Change 'your_dataset.csv' to the actual name of your file in the data/tabular folder!
df = pd.read_csv('../data/tabular/crop_yield.csv')

# Show the first 5 rows
print("Here is what your data looks like:")
display(df.head())

# Show a simple correlation heatmap
plt.figure(figsize=(8, 5))
# We only look at numbers, not text columns
numeric_data = df.select_dtypes(include=['number'])
sns.heatmap(numeric_data.corr(), annot=True, cmap='Blues')
plt.title("Correlation: Weather vs Crop Yield")
plt.show()
```
